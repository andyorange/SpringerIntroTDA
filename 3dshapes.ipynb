```python
alle_meine_entchen = []
```

```python
r1 = 8
r2 = 2
```

```python
ellipse_samples = []
```

```python
x = 0
y = 0
z = 0
rx = 8
ry = 6
rz = 3
```

```python
for i in range(1000):
    X = 1
    Y = 1
    Z = 1
    while X**2 + Y**2 + Z**2 > 1:
        X = uniform(-1, 1)
        Y = uniform(-1, 1)
        Z = uniform(-1, 1)
        
    N = sqrt(X**2 + Y**2 + Z**2)
    ellipse_samples.append([x + rx * X/N, y + ry * Y/N, z + rz * Z/N])
```

```python
sphere_samples = []
```

```python
x = r1/2 + 2.5*r2
y = r2 + r2/2
z = r2 + r2/4
r = 1.5*r2
```

```python
for i in range(200):
    X = r
    Y = r
    Z = r
    while X**2 + Y**2 + Z**2 > r**2:
        X = uniform(-r, r)
        Y = uniform(-r, r)
        Z = uniform(-r, r)
        
    N = sqrt(X**2 + Y**2 + Z**2)
    sphere_samples.append([x + r*X/N, y + r*Y/N, z + r*Z/N])
```

```python
alle_meine_entchen.append(np.array(ellipse_samples + sphere_samples))
```
