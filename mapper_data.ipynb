```python
x1 = [30*cos(angle) + x for angle, x, y in generators1]
y1 = [30*sin(angle) + y for angle, x, y in generators1]
x2 = [105*cos(angle) + x + 100 for angle, x, y in generators2]
y2 = [105*sin(angle) + y + 100 for angle, x, y in generators2]
```

```python
c1 = [round(6*(x+30)/300) for x, y in zip(x1, y1)]
c2 = [round(6*(x+30)/300) for x, y in zip(x2, y2)]
c1_2 = [round(6*(x+55)/300) for x, y in zip(x1, y1)]
c2_2 = [round(6*(x+55)/300) for x, y in zip(x2, y2)]
```

```python
with open("mapper_data.dat", "wt") as ofile:
    for x, y, c_1, c_2 in zip(x1, y1, c1, c1_2):
        ofile.write("%f,%f,-50,1,%d,%d\n"%(x, y, c_1, c_2))
    for x, y, c_1, c_2 in zip(x2, y2, c2, c2_2):
        ofile.write("%f,%f,-50,1,%d,%d\n"%(x, y, c_1, c_2))        
```
